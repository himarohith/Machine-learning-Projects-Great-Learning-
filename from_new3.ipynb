```python
size = 10
threshold = 0.5
df_corr = []

for i in range(0,size): 
    for j in range(i+1,size): 
        if (corrmat.iloc[i,j] >= threshold and corrmat.iloc[i,j] < 1) or (corrmat.iloc[i,j] < 0 and corrmat.iloc[i,j] <= -threshold):
            df_corr.append([corrmat.iloc[i,j],i,j])
```

```python
cols = df.columns
corr_list = sorted(df_corr,key=lambda x: -abs(x[0]))

for v,i,j in corr_list:
    print ("%s and %s = %.2f" % (cols[i],cols[j],v))
```

```python
for v,i,j in corr_list:
    sns.pairplot(df, hue='Cover_Type', size= 6, x_vars=cols[i], y_vars=cols[j])
    plt.show()
```
